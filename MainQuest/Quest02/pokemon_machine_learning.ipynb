```python
# 데이터셋을 pandas로 불러오기
import os

csv_path = '/content/drive/MyDrive/AIFFEL_2024/data/Pokemon.csv'
original_data = pd.read_csv(csv_path)
```

# **(1) 기본 모델이 동작하는 코드**

```python
# 원본 데이터 확인하기 - 포켓몬 800마리 및 각 포켓몬의 13가지 특성
print(original_data.shape)
original_data.head()
```

```python
# 원본 데이터의 컬럼 확인하기 - 13가지 특성
original_data.columns
```

```python
# 의미 없는 컬럼 #, 문자열을 포함하는 컬럼 (Name, Type1, Type2), 타겟 데이터터인 Legendary 제외하기
# 나머지는 배열로 features 변수에 담기
features = ['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation']
```

```python
# 타켓 데이터인 Legendary를 target 변수에 담기
target = 'Legendary'
```

```python
# features의 컬럼들에 해당하는 데이터를 변수 X에 저장하고 확인하기 - 포켓몬 800마리 및 각 포켓몬의 8가지 특성
X = original_data[features]
print(X.shape)
X.head()
```

```python
# target의 컬럼에 해당하는 데이터를 변수 y에 저장하고 확인하기 - 총 800마리 포켓몬 - 전설의 포켓몬 또는 일반 포켓문
y = original_data[target]
print(y.shape)
y.head()
```

```python
# 훈련 데이터와 테스트 데이터 분리하기 - 훈련 데이터 640 - 테스트 데이터 160
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# 의사결정 트리 모델을 변수 model에 저장하기
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=25)
model
```

```python
# 훈련 데이터로 모델 훈련시키기
model.fit(X_train, y_train)

# 테스트 데이터를 통해 예측값 얻기
y_pred = model.predict(X_test)
```

```python
# 예측값 (y_pred) 및 정답 (y_test) 확인하기
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

# **(2) 데이터 전처리**

```python
# 원본 데이터를 복사하여 pokemon 변수에 저장하고, pandas로 데이터 확인하기 - 포켓몬 800마리 및 각 포켓몬의 13가지 특성
pokemon = original_data.copy()
print(pokemon.shape)
pokemon.head()
```

```python
# "전설의 포켓몬" 데이터셋 확인하기 - 전설의 포켓몬 65마리 및 각 포켓몬의 13가지 특성
legendary = pokemon[pokemon["Legendary"] == True].reset_index(drop=True)
print(legendary.shape)
legendary.head()
```

```python
# 일반 포켓몬 데이터셋 확인하기 - 일반 포켓몬 735마리 및 각 포켓몬의 13가지 특성
ordinary = pokemon[pokemon["Legendary"] == False].reset_index(drop=True)
print(ordinary.shape)
ordinary.head()
```

```python
# 전체 컬럼 이해하기 - 13개의 컬럼 - 13가지 특성
print(len(pokemon.columns))
pokemon.columns
```

```python
# 컬럼 name_count 추가하기 - 이름의 길이 확인
pokemon["name_count"] = pokemon["Name"].apply(lambda i: len(i))
pokemon.head()
```

```python
# 컬럼 long_name 추가하기 - 이름의 길이가 10이상이면 True, 미만이면 False
pokemon["long_name"] = pokemon["name_count"] >= 10
pokemon.head()
```

```python
# 컬럼 Name_nospace 추가하기 - 띄어쓰기 없이 이름의 길이 확인
pokemon["Name_nospace"] = pokemon["Name"].apply(lambda i: i.replace(" ", ""))
pokemon.tail()
```

```python
# 컬럼 name_isalpha 추가하기 - 이름이 알파벳으로만 이루어졌는지 확인
pokemon["name_isalpha"] = pokemon["Name_nospace"].apply(lambda i: i.isalpha())
pokemon.head()
```

```python
# 이름에 알파벳이 아닌 다른 문자가 포함된 포켓몬 확인하기 - 포켓몬 9마리
print(pokemon[pokemon["name_isalpha"] == False].shape)
pokemon[pokemon["name_isalpha"] == False]
```

```python
# 해당 9마리 포켓몬의 이름 변경하기 - 알파벳이 아닌 것을 포함하는 이름 변경
pokemon = pokemon.replace(to_replace="Nidoran♀", value="Nidoran X")
pokemon = pokemon.replace(to_replace="Nidoran♂", value="Nidoran Y")
pokemon = pokemon.replace(to_replace="Farfetch'd", value="Farfetchd")
pokemon = pokemon.replace(to_replace="Mr. Mime", value="Mr Mime")
pokemon = pokemon.replace(to_replace="Porygon2", value="Porygon Two")
pokemon = pokemon.replace(to_replace="Ho-oh", value="Ho Oh")
pokemon = pokemon.replace(to_replace="Mime Jr.", value="Mime Jr")
pokemon = pokemon.replace(to_replace="Porygon-Z", value="Porygon Z")
pokemon = pokemon.replace(to_replace="Zygarde50% Forme", value="Zygarde Forme")

pokemon.loc[[34, 37, 90, 131, 252, 270, 487, 525, 794]]
```

```python
# 변경된 컬럼 Name으로 컬럼 Name_nospace를 업데이트하고, 알파벳이 아닌 것을 포함하는 이름 학인하기
pokemon["Name_nospace"] = pokemon["Name"].apply(lambda i: i.replace(" ", ""))

pokemon[pokemon["name_isalpha"] == False]
```

```python
import re

# 토큰화 함수 만들기
def tokenize(name):
    name_split = name.split(" ")
    tokens = []
    for part_name in name_split:
        a = re.findall('[A-Z][a-z]*', part_name)
        tokens.extend(a)

    return np.array(tokens)
```

```python
# 전설의 포켓몬의 이름에 사용된 토큰 확인하기 - 65개의 토큰
all_tokens = list(legendary["Name"].apply(tokenize).values)

token_set = []
for token in all_tokens:
    token_set.extend(token)

print(len(set(token_set)))
print(token_set)
```

```python
from collections import Counter

# 가장 많이 쓰인 토큰 확인하기
most_common = Counter(token_set).most_common(10)
most_common
```

```python
# 가장 많이 쓰인 토큰 10개 컬럼 추가하기 - 전설의 포켓몬 이름에 등장하는 토큰이 포켓몬 이름에 있는지 확인
for token, _ in most_common:
    # pokemon[token] = ... 형식으로 사용하면 뒤에서 warning이 발생합니다
    pokemon[f"{token}"] = pokemon["Name"].str.contains(token)

pokemon.head(10)
```

```python
# Type1 및 Type2 종류 수 확인하기
len(list(set(pokemon["Type 1"]))), len(list(set(pokemon["Type 2"])))
```

```python
# Type1과 Type2의 차이 확인 - NaN 데이터 외 나머지 18가지 속성이 동일함
set(pokemon["Type 2"]) - set(pokemon["Type 1"])
```

```python
# 포켓몬들의 모든 Type 속성을 변수 types에 담기
types = list(set(pokemon["Type 1"]))
print(len(types))
print(types)
```

```python
# 원-핫 인코딩 작업하기 - types에 담긴 18개의 속성 (컬럼)
for t in types:
    pokemon[t] = (pokemon["Type 1"] == t) | (pokemon["Type 2"] == t)

pokemon[[["Type 1", "Type 2"] + types][0]].head()
```

    - 불균형 데이터에서도 강력한 성능     

- SVM (Stochastic Gradient Descent)     
    - 대규모 데이터셋에 적합
    - 간단하고 빠름       
    - 적절한 하이퍼파라미터 튜닝 필요       
    - 데이터가 불균형하면 성능 저하        

- Logistic Regression.    
    - 이진 분류 문제에서 강력한 성능      
    - 선형 모델       
    - 데이터가 선형적으로 분리 가능하면 성능 좋음       
    - 불균형 데이터에 대해서는 성능이 좋지 않음         
    - 클래스 가중치 조정 가능        


            정밀도: 모델이 전설의 포켓몬이라고 예측한 것 중에서 실제로 전설의 포켓몬인 비율     

            재현율: 실제 전설의 포켓몬 중 모델이 정확하게 예측한 비율     

            F1-score: 정밀도와 재현율의 조화 평균으로, 불균형 데이터에서 중요한 평가 지표
